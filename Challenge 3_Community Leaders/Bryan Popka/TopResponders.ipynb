```
/* Overall Top Responders */
SELECT		TOP 100 ResponseLanguage
			,ResponseUserID
			,COUNT(*)
FROM		ADLMKT_Teradata_Cares.ProducersDirect_OFS
GROUP BY	1,2
ORDER BY	3 DESC;
```

```
/* Top 10 Responders By Language */
SELECT		ResponseLanguage
			,ResponseUserID
			,RespCnt
FROM		(SELECT		ResponseLanguage
						,ResponseUserID
						,COUNT(*) RespCnt
						,RANK() OVER (PARTITION BY ResponseLanguage ORDER BY RespCnt DESC) RespRank
			FROM		ADLMKT_Teradata_Cares.ProducersDirect_OFS
			GROUP BY	1,2) a
WHERE		RespRank <= 10
ORDER BY	1,3 ASC;
```

```
/* Most common relationships between Response Users and Question Users */
SELECT      TOP 100 ResponseUserID
            ,QuestionUserID
            ,COUNT(*)
FROM        ADLMKT_Teradata_Cares.ProducersDirect_OFS
GROUP BY    1,2
ORDER By    3 DESC;
```

```
/* The Responders with the most distinct Questioners that they have responded to */
SELECT      TOP 100 ResponseUserID
            ,COUNT(DISTINCT QuestionUserID) AS CountofQuestioners
FROM        ADLMKT_Teradata_Cares.ProducersDirect_OFS
GROUP BY    1
ORDER By    2 DESC;
```
