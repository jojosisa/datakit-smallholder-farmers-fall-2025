```
/* I used VSCode CoPilot using Claude Sonnet 4.5 to produce a list of the most common inclement weather terms found in the English language records. */

SELECT TOP 20 term, frequency FROM (
  SELECT CAST('drought' AS VARCHAR(50)) AS term, SUM(CASE WHEN LOWER(ResponseContent) LIKE '%drought%' THEN 1 ELSE 0 END) AS frequency FROM ADLMKT_Teradata_Cares.ProducersDirect_OFS WHERE ResponseLanguage='eng'
  UNION ALL SELECT 'flood', SUM(CASE WHEN LOWER(ResponseContent) LIKE '%flood%' THEN 1 ELSE 0 END) FROM ADLMKT_Teradata_Cares.ProducersDirect_OFS WHERE ResponseLanguage='eng'
  UNION ALL SELECT 'storm', SUM(CASE WHEN LOWER(ResponseContent) LIKE '%storm%' THEN 1 ELSE 0 END) FROM ADLMKT_Teradata_Cares.ProducersDirect_OFS WHERE ResponseLanguage='eng'
  UNION ALL SELECT 'frost', SUM(CASE WHEN LOWER(ResponseContent) LIKE '%frost%' THEN 1 ELSE 0 END) FROM ADLMKT_Teradata_Cares.ProducersDirect_OFS WHERE ResponseLanguage='eng'
  UNION ALL SELECT 'hail', SUM(CASE WHEN LOWER(ResponseContent) LIKE '%hail%' THEN 1 ELSE 0 END) FROM ADLMKT_Teradata_Cares.ProducersDirect_OFS WHERE ResponseLanguage='eng'
  UNION ALL SELECT 'hurricane', SUM(CASE WHEN LOWER(ResponseContent) LIKE '%hurricane%' THEN 1 ELSE 0 END) FROM ADLMKT_Teradata_Cares.ProducersDirect_OFS WHERE ResponseLanguage='eng'
  UNION ALL SELECT 'tornado', SUM(CASE WHEN LOWER(ResponseContent) LIKE '%tornado%' THEN 1 ELSE 0 END) FROM ADLMKT_Teradata_Cares.ProducersDirect_OFS WHERE ResponseLanguage='eng'
  UNION ALL SELECT 'lightning', SUM(CASE WHEN LOWER(ResponseContent) LIKE '%lightning%' THEN 1 ELSE 0 END) FROM ADLMKT_Teradata_Cares.ProducersDirect_OFS WHERE ResponseLanguage='eng'
  UNION ALL SELECT 'landslide', SUM(CASE WHEN LOWER(ResponseContent) LIKE '%landslide%' THEN 1 ELSE 0 END) FROM ADLMKT_Teradata_Cares.ProducersDirect_OFS WHERE ResponseLanguage='eng'
  UNION ALL SELECT 'erosion', SUM(CASE WHEN LOWER(ResponseContent) LIKE '%erosion%' THEN 1 ELSE 0 END) FROM ADLMKT_Teradata_Cares.ProducersDirect_OFS WHERE ResponseLanguage='eng'
  UNION ALL SELECT 'typhoon', SUM(CASE WHEN LOWER(ResponseContent) LIKE '%typhoon%' THEN 1 ELSE 0 END) FROM ADLMKT_Teradata_Cares.ProducersDirect_OFS WHERE ResponseLanguage='eng'
  UNION ALL SELECT 'cyclone', SUM(CASE WHEN LOWER(ResponseContent) LIKE '%cyclone%' THEN 1 ELSE 0 END) FROM ADLMKT_Teradata_Cares.ProducersDirect_OFS WHERE ResponseLanguage='eng'
  UNION ALL SELECT 'blizzard', SUM(CASE WHEN LOWER(ResponseContent) LIKE '%blizzard%' THEN 1 ELSE 0 END) FROM ADLMKT_Teradata_Cares.ProducersDirect_OFS WHERE ResponseLanguage='eng'
  UNION ALL SELECT 'avalanche', SUM(CASE WHEN LOWER(ResponseContent) LIKE '%avalanche%' THEN 1 ELSE 0 END) FROM ADLMKT_Teradata_Cares.ProducersDirect_OFS WHERE ResponseLanguage='eng'
  UNION ALL SELECT 'wildfire', SUM(CASE WHEN LOWER(ResponseContent) LIKE '%wildfire%' THEN 1 ELSE 0 END) FROM ADLMKT_Teradata_Cares.ProducersDirect_OFS WHERE ResponseLanguage='eng'
  UNION ALL SELECT 'earthquake', SUM(CASE WHEN LOWER(ResponseContent) LIKE '%earthquake%' THEN 1 ELSE 0 END) FROM ADLMKT_Teradata_Cares.ProducersDirect_OFS WHERE ResponseLanguage='eng'
  UNION ALL SELECT 'volcano', SUM(CASE WHEN LOWER(ResponseContent) LIKE '%volcano%' THEN 1 ELSE 0 END) FROM ADLMKT_Teradata_Cares.ProducersDirect_OFS WHERE ResponseLanguage='eng'
  UNION ALL SELECT 'tsunami', SUM(CASE WHEN LOWER(ResponseContent) LIKE '%tsunami%' THEN 1 ELSE 0 END) FROM ADLMKT_Teradata_Cares.ProducersDirect_OFS WHERE ResponseLanguage='eng'
  UNION ALL SELECT 'fog', SUM(CASE WHEN LOWER(ResponseContent) LIKE '%fog%' THEN 1 ELSE 0 END) FROM ADLMKT_Teradata_Cares.ProducersDirect_OFS WHERE ResponseLanguage='eng'
  UNION ALL SELECT 'heatwave', SUM(CASE WHEN LOWER(ResponseContent) LIKE '%heatwave%' THEN 1 ELSE 0 END) FROM ADLMKT_Teradata_Cares.ProducersDirect_OFS WHERE ResponseLanguage='eng'
) AS counts
ORDER BY frequency DESC
```

```
/* Using the list of terms produced above, I then had CoPilot translate those terms into Swahili, Lagunda, and Nyn
to use as part of the IN-List for the following queries that I wrote myself.

Get a total count of the messages by day vs. those messages that contain severe weather terms */
SELECT      CommDt
            ,SUM(TotalCount) AS TotalCnt
            ,SUM(SevWeatherCount) AS SevWeatherCnt
FROM        (SELECT      CAST(QuestionSentDt AS DATE) AS CommDt
                        ,COUNT(*) AS TotalCount
                        ,SUM(CASE WHEN QuestionContent LIKE ANY ('%erosion%', '%drought%', '%flood%', '%frost%', '%cyclone%', '%hail%', '%fog%', '%storm%', '%volcano%', '%hurricane%', '%landslide%', '%landslide%', '%earthquake%', '%tornado%', '%avalanche%', '%typhoon%', '%heatwave%', '%blizzard%', '%wildfire%', '%tsunami%') THEN 1 ELSE 0 END) AS SevWeatherCount
            FROM        ADLMKT_Teradata_Cares.ProducersDirectMod_OFS
            WHERE       QuestionLanguage = 'eng'
            GROUP BY    1
            UNION ALL
            SELECT      CAST(ResponseSentDt AS DATE) AS CommDt
                        ,COUNT(*) AS TotalCount
                        ,SUM(CASE WHEN ResponseContent LIKE ANY ('%erosion%', '%drought%', '%flood%', '%frost%', '%cyclone%', '%hail%', '%fog%', '%storm%', '%volcano%', '%hurricane%', '%landslide%', '%landslide%', '%earthquake%', '%tornado%', '%avalanche%', '%typhoon%', '%heatwave%', '%blizzard%', '%wildfire%', '%tsunami%') THEN 1 ELSE 0 END) AS SevWeatherCount
            FROM        ADLMKT_Teradata_Cares.ProducersDirectMod_OFS
            WHERE       QuestionLanguage = 'eng'
            GROUP BY    1) AS a
GROUP BY    CommDt
ORDER BY    CommDt ASC;
```

```
/* Creates a timeseries analysis showing total messages per day vs. those mentioning inclement weather terms 
I divide the total message count by 100 just so that it and the SevWeatherCount are on the same scale
for charting purposes.
*/
SELECT      CommDt
            ,CntType
            ,SUM(MessageCount) AS MessageCnt
FROM        (SELECT      CAST(QuestionSentDt AS DATE) AS CommDt
                        ,CAST('TotalCount' AS VARCHAR(30)) AS CntType
                        ,COUNT(*)/100 AS MessageCount
            FROM        ADLMKT_Teradata_Cares.ProducersDirectMod_OFS
            -- WHERE       QuestionLanguage = 'eng'
            GROUP BY    1
            UNION ALL
            SELECT      CAST(QuestionSentDt AS DATE) AS CommDt
                        ,CAST('SevWeatherCount' AS VARCHAR(30)) AS CntType
                        ,SUM(CASE WHEN QuestionContent LIKE ANY ('%erosion%', '%drought%', '%flood%', '%frost%', '%cyclone%', '%hail%', '%fog%', '%storm%', '%volcano%', '%hurricane%', '%landslide%', '%landslide%', '%earthquake%', '%tornado%', '%avalanche%', '%typhoon%', '%heatwave%', '%blizzard%', '%wildfire%', '%tsunami%') THEN 1 ELSE 0 END) AS MessageCount
            FROM        ADLMKT_Teradata_Cares.ProducersDirectMod_OFS
            WHERE       QuestionLanguage = 'eng'
            GROUP BY    1
            UNION ALL
            SELECT      CAST(QuestionSentDt AS DATE) AS CommDt
                        ,CAST('SevWeatherCount' AS VARCHAR(30)) AS CntType
                        ,SUM(CASE WHEN QuestionContent LIKE ANY ('%mmomonyoko%', '%ukame%', '%mafuriko%', '%baridi kali%', '%kimbunga%', '%mvua ya mawe%', '%ukungu%', '%dhoruba%', '%volkano%', '%kimbunga%', '%maporomoko ya ardhi%', '%radi%', '%tetemeko la ardhi%', '%tufani%', '%maporomoko ya theluji%', '%kimbunga%', '%joto kali%', '%dhoruba ya theluji%', '%moto wa porini%', '%tsunami') THEN 1 ELSE 0 END) AS MessageCount
            FROM        ADLMKT_Teradata_Cares.ProducersDirectMod_OFS
            WHERE       QuestionLanguage = 'swa'
            GROUP BY    1
            UNION ALL
            SELECT      CAST(QuestionSentDt AS DATE) AS CommDt
                        ,CAST('SevWeatherCount' AS VARCHAR(30)) AS CntType
                        ,SUM(CASE WHEN QuestionContent LIKE ANY ('okukuba kwa nkondo%', '%okwetaagisa amajji%', '%amazzi amangi%', '%okuffu%', '%ensi eyokya%', '%ssente z''amasannyalaze%', '%omusolo%', '%dhoruba%', '%volkano%', '%ensi eyokya%', '%maporomoko ga nsi%', '%ekiyingiza%', '%tetemeko lya nsi%', '%tufani%', '%maporomoko ga snow%', '%kimbunga%', '%joto kali%', '%dhoruba ya snow%', '%moto gwa miti%', '%tsunami') THEN 1 ELSE 0 END) AS MessageCount
            FROM        ADLMKT_Teradata_Cares.ProducersDirectMod_OFS
            WHERE       QuestionLanguage = 'lug'
            GROUP BY    1
            UNION ALL
            SELECT      CAST(QuestionSentDt AS DATE) AS CommDt
                        ,CAST('SevWeatherCount' AS VARCHAR(30)) AS CntType
                        ,SUM(CASE WHEN QuestionContent LIKE ANY ('okukuba kwa nkondo%', '%okwetaagisa amajji%', '%amazzi amangi%', '%okuffu%', '%ensi eyokya%', '%ssente z''amasannyalaze%', '%omusolo%', '%dhoruba%', '%volkano%', '%ensi eyokya%', '%maporomoko ga nsi%', '%ekiyingiza%', '%tetemeko lya nsi%', '%tufani%', '%maporomoko ga snow%', '%kimbunga%', '%joto kali%', '%dhoruba ya snow%', '%moto gwa miti%', '%tsunami') THEN 1 ELSE 0 END) AS MessageCount
            FROM        ADLMKT_Teradata_Cares.ProducersDirectMod_OFS
            WHERE       QuestionLanguage = 'nyn'
            GROUP BY    1) AS a
GROUP BY    CommDt,CntType
ORDER BY    CommDt;
```
