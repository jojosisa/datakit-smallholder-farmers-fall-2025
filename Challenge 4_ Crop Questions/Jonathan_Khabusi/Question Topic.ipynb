```python
#df.loc[df['response_language'] == 'spa']
#filtering all responses in Spanish
engquestions = df.loc[df['question_language'] == 'eng']
engquestions.drop_duplicates(subset=['question_content'], inplace=True)
print(engquestions.head(5))
```

```python
#PART 1
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import HashingVectorizer
from sklearn.cluster import MiniBatchKMeans
from sklearn.preprocessing import Normalizer
from sklearn.decomposition import TruncatedSVD
import matplotlib.pyplot as plt
import seaborn as sns

# --- CONFIGURATION ---
CHUNK_SIZE = 10000
total_rows = len(engquestions)
n_batches = int(np.ceil(total_rows / CHUNK_SIZE))
TRAIN_LIMIT_BATCHES = 50  # Train on first 50 batches (500k rows) only. 
                          # 500k is usually enough to stabilize clusters.

# 1. SETUP PIPELINE
vectorizer = HashingVectorizer(n_features=2**18, analyzer='char_wb', ngram_range=(3, 5), norm=None)
normalizer = Normalizer()
kmeans = MiniBatchKMeans(n_clusters=50, random_state=42, batch_size=CHUNK_SIZE)

# List to store results
results_list = []

print(f"Processing {total_rows} rows...")

# --- THE SINGLE EFFICIENT LOOP ---
for i in range(n_batches):
    # A. Get Data Chunk
    start = i * CHUNK_SIZE
    end = start + CHUNK_SIZE
    chunk = engquestions['question_content'].iloc[start:end].dropna().astype(str)
    
    if chunk.empty:
        continue

    # B. Transform
    X_batch = vectorizer.transform(chunk)
    X_batch = normalizer.transform(X_batch)
    
    # C. Train or Predict?
    # We only train on the first few batches to save time.
    # Once the model is "smart" enough, we stop training and just predict.
    if i < TRAIN_LIMIT_BATCHES:
        kmeans.partial_fit(X_batch)
        
    # D. Predict Labels (Assign Cluster ID)
    batch_labels = kmeans.predict(X_batch)
    
    # E. Store Results (Keep index so we can map back to original dataframe)
    # We store a mini-dataframe for this batch
    batch_results = pd.DataFrame({
        'original_index': chunk.index,
        'cluster_label': batch_labels,
        'question_text': chunk.values # Optional: Keep text if you have RAM
    })
    results_list.append(batch_results)
    
    if i % 10 == 0:
        print(f"Processed batch {i}/{n_batches} - Mode: {'Training' if i < TRAIN_LIMIT_BATCHES else 'Labeling Only'}")

print("Merging results...")
final_df = pd.concat(results_list, ignore_index=True)
print("Done! 'final_df' now has your clusters.")
```

```python
#PART 2
print("Generating Visualization...")

# 1. Take a random sample of 5,000 rows (enough to see patterns)
# We use the final_df created above
sample_df = final_df.sample(n=5000, random_state=42)

# 2. Re-vectorize this small sample for plotting
# (We need the vector form to calculate X/Y coordinates)
X_sample = vectorizer.transform(sample_df['question_text'])
X_sample = normalizer.transform(X_sample)

# 3. Reduce Dimensions to 2D using SVD
svd = TruncatedSVD(n_components=2, random_state=42)
coords = svd.fit_transform(X_sample)

# 4. Plot
plt.figure(figsize=(12, 8))
scatter = plt.scatter(coords[:, 0], coords[:, 1], c=sample_df['cluster_label'], cmap='tab20', alpha=0.6, s=10)
plt.colorbar(scatter, label='Cluster ID')
plt.title('Farmer Questions Clusters (2D Projection)')
plt.xlabel('SVD Component 1')
plt.ylabel('SVD Component 2')
plt.show() 
```

```python
import pandas as pd

# Define the number of example questions for each cluster
QUESTIONS_PER_CLUSTER = 5

print("Generating representative samples for export...")

# Create an empty list to store the representative samples from all clusters
representative_samples = []

# Loop through every unique cluster ID found in your results
for cluster_id in sorted(final_df['cluster_label'].unique()):
    
    # Filter the dataframe to get rows belonging to the current cluster
    cluster_data = final_df[final_df['cluster_label'] == cluster_id]
    
    # Check if the cluster has enough samples
    if len(cluster_data) >= QUESTIONS_PER_CLUSTER:
        # Randomly select the required number of questions from this cluster
        sample = cluster_data.sample(n=QUESTIONS_PER_CLUSTER, random_state=42)
    else:
        # If the cluster is too small, take all available questions
        sample = cluster_data
    
    # Add a descriptive column for manual labeling later
    sample.loc[:, 'Suggested_Cluster_Name'] = f"Cluster {cluster_id}: Name This Topic"
    
    # Append the sample to our main list
    representative_samples.append(sample)

# Combine all the samples into a single export dataframe
export_df = pd.concat(representative_samples)

# Reorder columns for easy viewing in Excel
export_df = export_df[['Suggested_Cluster_Name', 'cluster_label', 'question_text', 'original_index']]

# Export to an Excel file (Requires 'openpyxl' library: !pip install openpyxl)
export_df.to_excel("Cluster_Summary_for_Review.xlsx", index=False)

print("\n✅ Export complete! Look for 'Cluster_Summary_for_Review.xlsx' in your directory.")
print(f"File contains {len(export_df)} representative questions ready for manual naming.")
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from sklearn.decomposition import TruncatedSVD

# STEP 1: PREPARE A SAMPLE
# We cannot plot 2.7 million points. We take a random sample of 5,000.
# This is enough to show the structure without crashing the plotting library.
plot_sample = final_df.sample(n=5000, random_state=42).copy()

print("1. Vectorizing sample for visualization...")
# We must turn text into numbers (Vectors) again to get coordinates
X_plot = vectorizer.transform(plot_sample['question_text'])
X_plot = normalizer.transform(X_plot)

# STEP 2: CALCULATE COORDINATES (X and Y)
print("2. Reducing dimensions to X and Y...")
# TruncatedSVD squashes the complex data down to 2 dimensions (2D)
svd = TruncatedSVD(n_components=2, random_state=42)
coords = svd.fit_transform(X_plot)

# We save these X and Y coordinates into our dataframe so we can use them easily
plot_sample['x'] = coords[:, 0]
plot_sample['y'] = coords[:, 1]

# STEP 3: PLOT THE DOTS
print("3. Generating Plot...")
plt.figure(figsize=(14, 11)) # Make the image large (15x12 inches)

# Scatter plot: distinct colors for each cluster
# c=plot_sample['cluster_label'] colors them by their ID number
plt.scatter(plot_sample['x'], plot_sample['y'], 
            c=plot_sample['cluster_label'], 
            cmap='tab20', # A color palette with many distinct colors
            alpha=0.4,    # Transparency (0.4) lets you see overlapping dots
            s=10)         # Size of dots

# STEP 4: CALCULATE & ADD TEXT LABELS
# We group by 'Topic_Name' and find the average X and Y.
# This gives us the "center" of that topic cloud.

# Get a list of all unique topics in this sample
unique_topics = plot_sample['Topic_Name'].unique()

for topic in unique_topics:
    # Get all rows belonging to this specific topic
    topic_data = plot_sample[plot_sample['Topic_Name'] == topic]
    
    # Calculate the center (mean) position
    center_x = topic_data['x'].mean()
    center_y = topic_data['y'].mean()
    
    # Add the text label at that center position
    plt.text(x=center_x, 
             y=center_y, 
             s=topic, # The text to display (the name)
             fontsize=9, 
             fontweight='bold',
             color='black',
             # This box makes the text readable even on top of colored dots
             bbox=dict(facecolor='white', alpha=0.7, edgecolor='none', pad=1))

# Final Formatting
plt.title('Farmer Questions by Topic', fontsize=18)
plt.xlabel('Dimension 1')
plt.ylabel('Dimension 2')
plt.grid(True, alpha=0.3)

# Save the plot so you can download it
plt.savefig("labeled_cluster_map.png", dpi=150)
plt.show()
```

```python
extradata = engquestions[["question_user_country_code", "question_user_created_at"]]
#used right join to keep all rows from extradata
final_df_timeplace = final_df.merge(extradata, left_index=True, right_index=True, how='right')
#used format="mixed" to handle various date formats
final_df_timeplace["Month"], final_df_timeplace["Year"] = pd.to_datetime(final_df_timeplace["question_user_created_at"], format="mixed", errors='coerce').dt.month, pd.to_datetime(final_df_timeplace["question_user_created_at"], format="mixed", errors='coerce').dt.year
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
# A.1: Grouping the data
# We count how many times a topic appears in each month
#The .unstack() method pivots the data so that Months are rows and Topics are columns. 
# A heatmap allows you to instantly spot "hot" months for specific topics.
topic_trends = final_df_timeplace.groupby(['Month', 'Topic_Name']).size().unstack(fill_value=0)

# A.2: Sorting index
topic_trends = topic_trends.sort_index()

# A.3: Visualization (Heatmap)
plt.figure(figsize=(12, 6))
sns.heatmap(topic_trends, cmap="YlGnBu", annot=False)
plt.title('Intensity of Topics over Time')
plt.ylabel('Month')
plt.xlabel('Topic Name')
plt.show()
```

```python
# B.1: Grouping the data
country_dist = final_df_timeplace.groupby(['question_user_country_code', 'Topic_Name']).size().unstack(fill_value=0)

# B.2: Normalizing - converts raw counts to percentages so rows sum to 100%
country_dist_pct = country_dist.div(country_dist.sum(axis=1), axis=0) * 100

# B.3: Visualization (Stacked Bar Chart)
country_dist_pct.plot(kind='bar', stacked=True, figsize=(12, 6), colormap='tab20')

plt.title('Topic Distribution by Country')
plt.xlabel('Country')
plt.ylabel('Percentage of Discussion (%)')
plt.legend(title='Topic Name', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```
